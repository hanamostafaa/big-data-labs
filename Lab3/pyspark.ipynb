```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder\
    .master("local[*]")\
    .appName("CustomerData")\
    .getOrCreate()
```

```python
sc=spark.sparkContext
```

## USING RDD

```python
rdd = spark.read.csv(r"D:\in.csv", header=False).rdd
```

```python
kv_rdd = rdd.map(lambda row: (row[0], float(row[1])))
```

```python
reduced_rdd = kv_rdd.reduceByKey(lambda a, b: a + b)
```

```python
final_df = reduced_rdd.toDF(["ID", "val"])
```

```python
final_df.toPandas().to_csv(r"D:/output.csv", index=False, header=False)
```

## USING DATAFRAMES

```python
from pyspark.sql.functions import col, sum as spark_sum
```

```python
df = spark.read.csv(
    r"D:\in.csv"
)
```

```python
df = df.withColumnRenamed("_c0", "ID").withColumnRenamed("_c1", "val")
```

```python
df_reduced = df.groupBy("ID").agg(spark_sum("val").alias("total_val"))
```

```python
df_reduced.toPandas().to_csv("D:/lab3.csv", index=False)
```
